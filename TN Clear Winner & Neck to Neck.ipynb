## Clear Winner & Neck to Neck
Candiate who won with large margin and who won with smallest margin

```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
%matplotlib inline
plt.style.use('ggplot')
```

```python
indata = pd.read_csv("Data/Tamil Nadu.csv")
```

```python
indata.head()
```

```python
groups = indata.sort_values(['cname', 'Total Votes'], ascending=False).groupby(['cname'])
result = []
for idx, group in groups:
    result.append({
        'Constituency':idx,
        'Winner':group.iloc[0]['Party'],
        'Winner_votes':group.iloc[0]['Total Votes'],
        'Winner_candidate':group.iloc[0]['Candidate'],
        'Runner':group.iloc[1]['Party'],
        'Runner_votes':group.iloc[1]['Total Votes'],
        'Runner_candidate':group.iloc[1]['Candidate'],
        'Total_Votes':group['Total Votes'].sum()
    })
result = pd.DataFrame(result)
```

```python
result['VoteDiff'] = result['Winner_votes'] - result['Runner_votes']
result['PctDiff'] = (result['VoteDiff']) / result['Total_Votes']
```

### Clear Winner
